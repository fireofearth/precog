```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm
import matplotlib.gridspec as gridspec

N = 10
np.random.seed(42)

colors=cm.OrRd_r(np.linspace(.2, .6, N))

data = np.random.binomial(20, .3, (N, 100))

fig2, axs = plt.subplots(N, figsize=(10, 6))
for i, d in enumerate(data):
    axs[i].hist(d, range(20), color=colors[i], label=str(i))
fig2.legend(loc='upper center', ncol=5)
```

```python
def format_axes(fig):
    for i, ax in enumerate(fig.axes):
        ax.text(0.5, 0.5, "ax%d" % (i+1), va="center", ha="center")
        ax.tick_params(labelbottom=False, labelleft=False)


# gridspec inside gridspec
f = plt.figure(figsize=(20, 15))

gs = gridspec.GridSpec(4, 3, figure=f, wspace=0.08, hspace=0.08)
gs_x_histo = gs[0:2, 0].subgridspec(10, 1, hspace=0.04)
for k in range(10):
    f.add_subplot(gs_x_histo[k,0])
gs_y_histo = gs[2, 1:3].subgridspec(1, 10, wspace=0.04)
for k in range(10):
    f.add_subplot(gs_y_histo[0,k])
ax_scatter = f.add_subplot(gs[2,0])
ax_map     = f.add_subplot(gs[0:2,1:3])
ax_boxplot = f.add_subplot(gs[3,:])

format_axes(f)

plt.show()
```
